## Processing data

```python
#Creating a copy
data = raw_data.copy()
```

```python
# Checking null values
data.isnull().sum(axis=0)  
```

```python
# Filling missing values
data['reviewText']=data['reviewText'].fillna('Nil')
```

```python
# Checking null values
data.isnull().sum(axis=0)  
```

```python
# Removing html tags
from bs4 import BeautifulSoup

def strip_html(text):
    soup = BeautifulSoup(text, "html.parser")
    return soup.get_text()

data['reviewText'] = data['reviewText'].apply(lambda x: strip_html(x))
data.head(10)
```

```python
# removing numbers
import re
def remove_numbers(text):
  text = re.sub(r'\d+', '', text)
  return text

data['reviewText'] = data['reviewText'].apply(lambda x: remove_numbers(x))
data.head(10)
```

```python
# converting text to lower
data['reviewText'] = data.apply(lambda row: row['reviewText'].lower(), axis=1) 
```

```python
! pip install nltk
import nltk
from nltk.tokenize import word_tokenize 

nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')  
from nltk.corpus import stopwords 
stop = set(stopwords.words('english'))
```

```python
# Tokenization of data
data['reviewText'] = data.apply(lambda row: word_tokenize(row['reviewText']), axis=1) 
```

```python
# Removing non - ascii characters from the list
import unicodedata
def remove_non_ascii(words):
    new_words = []
    for word in words:
        new_word = unicodedata.normalize('NFKD', word).encode('ascii', 'ignore').decode('utf-8', 'ignore')
        new_words.append(new_word)
    return new_words

data['reviewText'] = data.apply(lambda row: remove_non_ascii(row['reviewText']), axis=1)
data.head(10)
```

```python
# removing punctuations and special characters

import re
def remove_punctuations(words):
    """Remove punctuations from list of tokenized words"""
    new_words = []
    for word in words:
        new_word = re.sub(r'[^\w\s]', '', word)
        if new_word != '':
            new_words.append(new_word)
    return new_words


data['reviewText'] = data.apply(lambda row: remove_punctuations(row['reviewText']), axis=1)
data.head(10)
```

```python
# remove stop words
def remove_stopwords(words):
    """Remove stop words from list of tokenized words"""
    new_words = []
    for word in words:
        if word not in stop:
            new_words.append(word)
    return new_words

data['reviewText'] = data.apply(lambda row: remove_stopwords(row['reviewText']), axis=1)
data.head(10)
```

```python
from nltk.stem.wordnet import WordNetLemmatizer         # Import Lemmatizer.
lemmatizer = WordNetLemmatizer()
def lemmatize_list(words):
    new_words = []
    for word in words:
        new_words.append(lemmatizer.lemmatize(word, pos='v'))
    return new_words

data['reviewText'] = data.apply(lambda row: lemmatize_list(row['reviewText']), axis=1)
data.head(10)
```

```python
# Creating sub dataframe
selected_columns = data[["reviewText", "overall"]]
review_rating_data = selected_columns.copy()
review_rating_data.head()
```

```python
# converting lists in reviewText to strings
review_rating_data['reviewText'] = review_rating_data.apply(lambda row: ' '.join(row['reviewText']), axis=1)
review_rating_data.head()
```

```python
# removing duplicate rows in dataframe
review_rating_data.sort_values("reviewText", inplace = True)
review_rating_data.drop_duplicates(keep = "first", inplace = True)
review_rating_data.shape
```

## Count Vectorization

```python
# Vectorization (Convert text data to numbers).
from sklearn.feature_extraction.text import CountVectorizer

vectorizer = CountVectorizer(max_features=1000)                # Keep only 5000 features as number of features will increase the processing time.
data_features = vectorizer.fit_transform(review_rating_data['reviewText'])

data_features = data_features.toarray()                        # Convert the data features to array.
```

```python
data_features.shape
```

```python
X = data_features

y = review_rating_data.overall
```

## Training MultinomialNB data


```python
# Split data into training and testing set.

from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score, confusion_matrix

multinomial = MultinomialNB()

multinomial.fit(X_train, y_train)
prediction = multinomial.predict(X_test)

print("\nAccuracy is {}:".format(accuracy_score(prediction,y_test)))
matrix = confusion_matrix(prediction,y_test)
print("\nConfusion matrix is, \n")
print(matrix)
```

## Dependency Parse Tree

```python
# Building dependency tree
import spacy
nlp = spacy.load('en_core_web_sm')
 
text = 'When Jon Snow is stranded north of the Wall, half-frozen and under attack by wights, Benjen rides in and puts Jon on his horse.'

for token in nlp(text):
 print(token.text,'=>',token.dep_,'=>',token.head.text)
```

```python
# Displaying dependency parser tree
from spacy import displacy
displacy.render(nlp(text),jupyter=True)
```
